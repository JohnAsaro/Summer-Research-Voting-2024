Plotting voting results for different voting rules, comparing the results based on average theta coefficent

(normal distrubution)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import numpy as np


#mpl.rcParams.update(mpl.rcParamsDefault) #to reset style
style.use('ggplot')
style.use('tableau-colorblind10')
style.use('seaborn-v0_8-paper')
```

```python
df = pd.read_csv('voting_results_k_1_normal.csv')

print(df.columns)

average_theta_values = {
    'Plurality': df['Plurality θ'].mean(),
    'Borda': df['Borda θ'].mean(),
    'Copeland': df['Copeland θ'].mean(),
    'STV': df['STV θ'].mean(),
    'Greatest θ-Winning Set': df['Greatest θ Coefficient'].mean()
}
```

Plot average theta vaule for each voting rule on bar graph

```python
#Plotting
colors = ['skyblue', 'lightgreen', 'salmon', 'orange', 'lightseagreen']
plt.figure(figsize=(10, 6))
bars = plt.bar(average_theta_values.keys(), average_theta_values.values(), color=colors)

#Adding a legend
#plt.legend(bars, average_theta_values.keys(), loc="center left", bbox_to_anchor=(1, 0.5))


#Adding titles and labels
plt.xlabel('Voting Rule', labelpad=15)  
plt.title('Average θ for Each Voting Rule across all different m values (k = 1/normal distrubution)', pad=20)
plt.ylabel('Average θ', labelpad=15)

for bar in bars: #Add the actual values on top of the bars
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 4), ha='center', va='bottom')


#Adjust the plot to fit the legend
plt.tight_layout()

#Show the plot
plt.show()
```

As m rises change in theta

```python
CB_color_cycle = ['skyblue', 'lightgreen', 'salmon', 'orange', 'lightseagreen']

#Create results dictionary
results = {
    "m": [3, 5, 10, 30, 50, 100],
    "Plurality θ": plurality_avg_all_m,
    "Borda θ": borda_avg_all_m,
    "Copeland θ": copeland_avg_all_m,
    "STV θ": stv_avg_all_m,
    "Greatest θ Coefficient": greatest_theta_winning_set_avg_all_m
}

#Extract data from results dictionary
m_values = results["m"]
plurality_theta = results["Plurality θ"]
borda_theta = results["Borda θ"]
copeland_theta = results["Copeland θ"]
stv_theta = results["STV θ"]
greatest_theta = results["Greatest θ Coefficient"]

#Plot the data
plt.figure(figsize=(10, 6))

plt.plot(m_values, plurality_theta, label='Plurality θ', marker='s', color=CB_color_cycle[0])
plt.plot(m_values, borda_theta, label='Borda θ', marker='o', color=CB_color_cycle[1])
plt.plot(m_values, copeland_theta, label='Copeland θ', marker='^', color=CB_color_cycle[2])
plt.plot(m_values, stv_theta, label='STV θ', marker='v', color=CB_color_cycle[3])
plt.plot(m_values, greatest_theta, label='Greatest θ Coefficient', marker = 'D', color=CB_color_cycle[4])

#Add labels and title
plt.xlabel('Number of Candidates (m)')
plt.ylabel('Theta (θ)')
plt.title('Change in average θ as m grows (k = 1/normal distrubution)')

#Set limits
plt.ylim(0.0, 1.025)

#Set x-axis ticks
plt.xticks(m_values)

#Add legend
plt.legend()

#Display the plot
plt.grid(True)
plt.show()
```
